# Baseline Models

Here, I will run some baseline models on the data. After splitting the data into train and test sets, I will run it through various model types to see which ones perform the best. Those that work best will be fine-tuned later.

```python
import pandas as pd
from scripts import print_metrics
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/cleaned_data.csv', index_col='id')
df
```

```python
y = df['status_group']
X = df.drop(['status_group'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=212)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(fit_intercept=False, C=1e12, solver='liblinear')
logreg.fit(X_train, y_train)
print_metrics(y_test, X_test, logreg)
```
